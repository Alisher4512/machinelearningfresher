**Importing the related libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Loading the Data**

```python
df1=pd.read_csv('/content/student_info.csv')
```

```python
df1.head()
```

```python
df1.shape
```

```python
df1.info()
```

```python
df=df1.fillna(df1.mean(),axis=0)
```

```python
df.describe()
```

**Visualizing the Data**

```python
sns.barplot(df['study_hours'],y=df['student_marks'])
sns.scatterplot(df['study_hours'],y=df['student_marks'])
sns.swarmplot(df['study_hours'],df['student_marks'])
```

```python
x=df.drop('study_hours',axis='columns')
x.shape
```

```python
y=df.drop('student_marks',axis='columns')
y.shape
```

**Importing related modules**

```python
from sklearn.model_selection import train_test_split,cross_val_score,RandomizedSearchCV,M_A_
```

**Splitting the Data**

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=30)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

**Training Regression Model**

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import RadiusNeighborsRegressor
from xgboost import XGBRegressor
from sklearn.preprocessing import StandardScaler
```

**Scaling the Data**

```python
sr=StandardScaler()
x_train_std=sr.fit_transform(x_train,y_train)
```

**Regressor Model**

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
y_predict=lr.predict(x_test)
y_predict
lr.score(x_test,y_test)
```

**Random_forest**

```python
lr=RandomForestRegressor()
lr.fit(x_train,y_train)
y_predict=lr.predict(x_test)
y_predict
lr.score(x_test,y_test)
```

**Decision Tree**

```python
lr=DecisionTreeRegressor()
lr.fit(x_train,y_train)
y_predict=lr.predict(x_test)
y_predict
lr.score(x_test,y_test)
```

**Support Vector Model**

```python
lr1=SVR()
lr1.fit(x_train,y_train)
y_predict=lr1.predict(x_test)
y_predict
lr1.score(x_test,y_test)
```

**Concatenation**

```python
pd.DataFrame(np.c_[x_test,y_test,y_predict],columns=['study_hours','original_marks','predicted_marks'])
```

```python
plt.scatter(x_train,y_train)
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test,y_predict,color='r')
```
